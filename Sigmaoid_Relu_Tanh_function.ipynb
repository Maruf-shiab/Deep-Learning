## Sigmoid Activation Function

The sigmoid function, also known as the logistic function, is a common activation function used in artificial neural networks. It maps any real-valued number into a value between 0 and 1.

**Mathematical Formula:**

$\sigma(x) = \frac{1}{1 + e^{-x}}$

**Key Characteristics:**

*   **Output Range:** The output of the sigmoid function is always between 0 and 1. This makes it suitable for binary classification problems where the output can be interpreted as a probability.
*   **Non-linearity:** The sigmoid function introduces non-linearity into the network, allowing it to learn complex relationships in the data. Without non-linear activation functions, a neural network, no matter how many layers it has, would essentially behave like a single-layer perceptron.
*   **Vanishing Gradients:** One of the main drawbacks of the sigmoid function is the vanishing gradient problem. For very large positive or negative inputs, the function's output becomes very close to 1 or 0, and the gradient in these regions is very small. This can make it difficult for the network to learn effectively during backpropagation, as the weights receive very small updates.
*   **Monotonic:** The function is monotonically increasing.
*   **Smooth:** It is a smooth, differentiable function, which is necessary for gradient-based optimization algorithms.

## ReLU (Rectified Linear Unit) Activation Function

The ReLU function is one of the most widely used activation functions in deep learning models, particularly in convolutional neural networks (CNNs) and deep neural networks.

**Mathematical Formula:**

$\text{ReLU}(x) = \max(0, x)$

This means that if the input `x` is positive, the output is `x`. If the input `x` is negative, the output is 0.

**Key Characteristics:**

*   **Non-linearity:** ReLU is a non-linear function, which is crucial for deep learning models to learn complex patterns and make them capable of solving complex tasks.
*   **Computational Efficiency:** Compared to sigmoid and tanh functions, ReLU is computationally very efficient. It involves simple operations like comparison with zero and returning the input value, making it faster to compute during forward and backward passes.
*   **Avoids Vanishing Gradient Problem (for positive inputs):** For positive inputs, the derivative of ReLU is 1, which helps to mitigate the vanishing gradient problem. This allows neural networks with many layers to train more effectively.
*   **Sparsity:** ReLU can lead to sparse activations. When the input to a ReLU unit is negative, the output is 0, effectively deactivating that neuron. This sparsity can be beneficial, as it can make the model more efficient and reduce overfitting.
*   **Dying ReLU Problem:** A potential drawback of ReLU is the 'dying ReLU' problem. If a neuron consistently receives negative inputs, its output will always be 0, and its gradient will also be 0. This means the neuron will stop learning (it 'dies') and will never activate again. Variants like Leaky ReLU or ELU were introduced to address this issue.

**Advantages over Sigmoid and Tanh:**

*   **Faster Convergence:** ReLU tends to achieve faster convergence during training compared to sigmoid and tanh due to its linear, non-saturating nature for positive inputs.
*   **Simpler Implementation:** Its simple mathematical form makes it easy to implement and debug.**bold text**

```python
def relu(x):
 return np.where(x>=0,x,0)
```

```python
x_relu_plot = np.linspace(-10, 10, 100)
y_relu_plot = relu(x_relu_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_relu_plot, y_relu_plot, label='ReLU Function')
plt.title('ReLU Function Plot')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.axhline(0, color='red', linestyle='--', linewidth=0.7, label='y=0')
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7, label='x=0')
plt.legend()
plt.show()
```

## Tanh (Hyperbolic Tangent) Activation Function

The Tanh (hyperbolic tangent) activation function is another popular choice, especially in recurrent neural networks. It is similar to the sigmoid function but produces outputs ranging from -1 to 1.

**Mathematical Formula:**

$\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

**Key Characteristics:**

*   **Output Range:** The output of the tanh function is centered at 0 and ranges from -1 to 1. This zero-centered property can be beneficial during optimization, as it helps in propagating gradients more effectively through the network.
*   **Non-linearity:** Like sigmoid, tanh is non-linear, allowing neural networks to learn complex patterns.
*   **Vanishing Gradients:** Tanh also suffers from the vanishing gradient problem, especially for very large positive or negative inputs, where the gradient becomes very close to zero.
*   **Steeper Gradients:** Compared to the sigmoid function, tanh has steeper gradients, which can sometimes lead to faster convergence during training.

**Comparison with Sigmoid:**

*   Tanh is preferred over sigmoid in many cases because its output is zero-centered, which can make the learning process easier for subsequent layers.

```python
def tanh(x):
  return np.tanh(x)

def tanh_derivative(x):
  return 1 - np.tanh(x)**2

x_values = np.linspace(-5, 5, 100)
y_tanh = tanh(x_values)
y_tanh_deriv = tanh_derivative(x_values)

plt.figure(figsize=(10, 7))
plt.plot(x_values, y_tanh, label='Tanh Function')
plt.plot(x_values, y_tanh_deriv, label='Tanh Derivative', linestyle='--')
plt.title('Tanh Function and its Derivative')
plt.xlabel('x')
plt.ylabel('Value')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
